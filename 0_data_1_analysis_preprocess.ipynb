# 数据预处理

```python
import pandas as pd
import os
data_dir = "./data/ml-latest-small"
ratings_df = pd.read_csv(os.path.join(data_dir, "ratings.csv"))
```

```python
import matplotlib.pyplot as plt
# 显示rating的userID和movieID的分布
plt.scatter(ratings_df['userId'], ratings_df['movieId'])
plt.xlabel('User ID')
plt.ylabel('Movie ID')
plt.title('User ID and Movie ID Distribution')
plt.show()
```

可以看到分布不连续，我们处理movieID成为连续的数值

```python
movieId = ratings_df.movieId.unique()
movieId
```

```python
movieId.sort()
movieId
```

```python
movieId_map = {movieId[i]: i+1 for i in range(len(movieId))}
```

```python
ratings_df['movieId'] = ratings_df['movieId'].map(movieId_map)
```

```python
# 重新显示rating的userID和movieID的分布
plt.scatter(ratings_df['userId'], ratings_df['movieId'])
plt.xlabel('User ID')
plt.ylabel('Movie ID')
plt.title('User ID and Movie ID Distribution')
plt.show()
```

可以看到现在变成稠密了

```python
ratings_df.to_csv(os.path.join(data_dir, "ratings_processed.csv"), index=False)
```
